## Heading
- 검색 엔진이 heading으로 구조 indexing함

```python
%%writefile test.html
<!doctype html>
<html lang="ko">
    <head>
        <meta charset="utf-8"/>
        <!-- 위의 meta tag만 title 위에 쓰고 나머지는 아래에 씀 -->
        <!-- 빠르게 문자열 세팅을 해주려고 -->
        <title>Simple HTML doc.</title>
        <meta name="viewport" content="width=device-width, initial-scale=1.0"/>
        <meta name="keywords" content="실습, 데이터진흥원, 수업"/>
        <!-- 위의 meta 키워드에서 content내의 내용이 바로 태그 -->
        <script>
            var car = {"make" : "Volkswagen"};
            car.year = '2003';
            console.log(car);
        </script>
        <style>
            * {margin:0;padding:0;}
            header {background-color:#f1f1f1;text-align:center;}
            header > h1 {padding:20px;}
            nav {overflow:hidden;background-color:#333;}
            nav li {display:block;float:left;padding:14px 16px;color:#f2f2f2;text-align:center;}
            nav li:hover {background-color:#ddd;color:black;}
            main > section {float:left;width:60%;}
            main > aside {float:right;width:30%;}
            main > section, main > aside {background-color:#aaa;margin-top:20px;padding:10px}
            main > section > article, main > aside > section {padding:20px;margin:20px;background-color:white;}
            article > section {background-color:#aaa;}
            p {padding:10px 0;text-align:center;border:1px solid #000;}
        </style>
    </head>
    <body>
        <main>
        <!-- main은 꽉 찬 화면, 중요하게 다룰 단위의 화면 잡을 때 쓴다 -->
            <header>
                <h1>Header</h1>
            </header>
            <nav>
                <ul>
                    <li>Home</li>
                    <li>News</li>
                    <li>Contact</li>
                    <li>About</li>
                </ul>
            </nav>
            
            <section>
                <article>
                    <h1>Title</h1>
                    <p>Content</p>
                </article>
                <article>
                    <h1>Title</h1>
                    <section>
                        <h1>Sub title1</h1>
                        <p>Sub content1</p>
                    </section>
                    <section>
                        <h1>Sub title2</h1>
                        <p>Sub content2</p>
                    </section>
                </article>
            </section>
            <aside>
                <section>
                    <h1>Advertisement</h1>
                    <p>Content</p>
                </section>
                <section>
                    <h1>Advertisement</h1>
                    <p>Content</p>
                </section>
            </aside>
            <footer>
            </footer>
        </main>

    </body>
</html>
```

## CSS
- Define styles for web pages
- Better type and layout controls
- Less work
- Potentially smaller documents
- Improved accessibility
- Reliable browser support


```python
%%writefile css.html
<!doctype html>
<html>
    <head>
        <style>
            div {background-color:rgba(255, 0, 0, .3);
                border:1px solid #aaaaaa;}
            #id1 {position:relative;height:50px;background-color:#bbbbbb;}
            #id2 {position:absolute;top:100px;left:200px;}
            #id3 {position:fixed;top:100px;left:200px;}
            #id {color:rgb(0, 0, 255);}
            #id1 > ul{position:absolute;top:50px;left:200px;border:1px solid #aaaaaa;display:none;}
            #id1:hover > ul{display:block;}
            .class {background-color:yellow;}
            div + p.class {background-color:green;}
            p.class {font-family:"맑은고딕", "Malgun Gothic", BatangChe, Sans-serif;
                    font-size: 2.0em;}
            /* 순서대로 먼저 적용되고, 없으면 다음 것이 적용됨 */
        </style>
    </head>
    <body>
        <!--
        <div id="id">Content1</div>
        <div class="class">Content2</div>
        <p class="class">Content3</p>
        -->
        <div id="id1">
            <ul>
                 <li>메뉴1</li>
                 <li>메뉴2</li>
            </ul>
                 
         </div>
        <div id="id2">Content2</div>
        <div id="id3">Content3</div>
    </body>
</html>
```

## Z-index
- 상대 위치(높이 개념)


```python
%%writefile div.html
<!dictype html>
<html>
    <head>
        <style>
            div {border:1px solid #aaaaaa;width:800px;height:200px;opacity:0.5;background-color:#aaaaaa;}
            #div2 {width:200px;height:500px;float:left;background-color:red;}
            #div3 {width:200;height:500px;float:right;background-color:blue;}
            #div4 {clear:both}
            /* 굳이 테이블로 안하고 clear로 float를 없애주면, float가 아닌 div1 아래에 붙던 애가 맨 아래로 붙게 됨 */
           #div1 + div {clear:both;content:"";display:table}
        </style>
    </head>
    <body>
        <div id="div1">Header</div>
        <div>
            <div id="div2">Side</div>
            <div id="div3">Content</div>
        </div>
        <div id="div4">Footer</div>
    </body>
</html>
```

```python
def bouncingBall(h, bounce, window):
    # your code
    if h == 0 or bounce == 0 or bounce == 1 or window >= h:
        return -1
    else:
        result = 1
        h *= bounce
        while h > window:
            result += 2
            h *= bounce
        return result
bouncingBall(30, 0.66, 1.5)
```
