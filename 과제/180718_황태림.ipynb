```python
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, ForeignKey

engine = create_engine("sqlite:///open_api_data.db", echo=True)
```

```python
Base = declarative_base()
```

```python
class OpenData(Base):
    __tablename__="OpenData"
    id = Column(Integer, primary_key=True)
    
    _returnType = Column(String)
    cityName = Column(String)
    cityNameEng = Column(String)
    coValue = Column(String)
    dataGubun = Column(String)
    dataTime = Column(String)
    districtCode = Column(String)
    districtNumSeq = Column(String)
    itemCode = Column(String)
    khaiValue = Column(String)
    no2Value = Column(String)
    numOfRows = Column(String)
    o3Value = Column(String)
    pageNo = Column(String)
    pm10Value = Column(String)
    pm25Value = Column(String)
    resultCode = Column(String)
    resultMsg = Column(String)
    searchCondition = Column(String)
    serviceKey = Column(String)
    sidoName = Column(String)
    so2Value = Column(String)
    totalCount = Column(String)
```

```python
Base.metadata.create_all(engine)
```
