## RDBMS 개념
- relation : table
- tuple : row 또는 record
- attribute : column 또는 field
- cardinality : number of rows
- degree : number of columns
- domain : legal values
- primary key : unique identifier

## 개체 관계 모델(Entity-Relationship Model, ERM)
- 개체(Entity) : 실 세계에 존재하는 실체 하나, 명사에 해당
- 관계(Relationship) : 개체 사이의 연관, 연결, 동사에 해당
 * 최소 대응수(min. cardinality)와 최대 대응수(max. cardinality) 구성
- 속성(Attribute) : 개체의 성질, 분류, 식별, 수량, 상태 등 나타내는 세부 항목
 * 관계 또한 속성을 가질 수 있다
- 기본키(Primary Key) : 개체의 고유 식별 가능한 속성
- 외부키(Foreign Key) : 다른 table의 식별키

## Mapping Cardinality
- 교수 : 학생 관계로 설명, 관계는 '지도'
- 교수는 꼭 학생을 지도해야 한다 : min. card. = 1
- 교수는 여러 명의 학생을 지도 가능하다 : max. card. = n
- 학생은 꼭 교수에게 지도를 받아야 한다 : min. card. = 1
- 학생은 여러 명의 교수에게 지도를 받을 수 없다 : max. card. = 1
- 위의 관계는 (교수, 지도) = (1, n) / (지도, 학생) = (1, 1)

## SQL
- Structured Query Language
- RDBMS 데이터를 관리하기 위한 언어
- 자료 검색, 관리, DB 스키마 생성, 수정, 객체 접근 조정 관리 등

## SQL 명령
- DDL(Data Definition Language)
 * CREATE, DROP(테이블 다 지움), ALTER, TRUNCATE(내용만 다 지움)...
- DML(Data Manipulation Language)
 * INSERT, UPDATE, DELETE, SELECT...
- DCL(Data Control Language)
 * GRANT, REVOKE...

## Data type
- Boolean(BOOLEAN) : 언어마다 명세가 다를 수 있어서 안쓰는 사람도 있음
- Character(CHAR, VARCHAR) : 길이 정형(CHAR), 길이 가변형(VARCHAR)라서 VARCHAR를 많이 씀
 * 날짜 같은 경우는 폭이 정해져 있으니 CHAR를 쓰기도 함(DATE를 더 많이 씀...)
 * CHAR : CHAR(5)면 3개 입력 시 나머지 2칸은 공백문자
- Exact numeric(NUMERIC, DECIMAL, INTEGER, SMALLINT, BIGINT) : 정수형
 * NUMERIC : NUMERIC(8, 2) 면 소수점 2자리라고 정하는 것
- Approximate numeric(REAL, FLOAT, DOUBLE) : 실수형
- Datetime(DATE, TIME, TIMESTAMP) : 시간
- Large object(CLOB, BLOB) : character string, binary string
 * CLOB : 엄청 큰 문자열 데이터 처리 시
 * BLOB : 위와 비슷한 목적이나 escape character 문제 등 해결 시 사용

## SQL 명령어(상세)
##### DDL
- CREATE TABLE {table name} ( {column definitions} ) {table parameters}
 * NOT NULL, UNIQUE, PRIMARY KEY(필수로 넣어준다), FOREIGN KEY, CHECK, DEFAULT(no value일 때의 default value), INDEX
 * ex> CREATE TABLE employees(id INTEGER PRIMARY KEY, first_name VARCHAR(50) not null);
- DROP {object type} {object name}
 * ex> DROP TABLE employees;
 - TRUNCATE {object type} {object name}
 - ALTER {object type} {object name} {parameters}
  * ex> ALTER TABLE employees ADD last_name VARCHAR(50);
 
##### DML
- INSERT INTO table_name(column 1, column2...) VALUES(value1, value2...)
- SELECT column1, column2... FROM table_name
 * WHERE(=[equal], <>[not equal], BETWEEN, LIKE[%a% 면 a 좌우로 아무거나], IN...), GROUP BY, HAVING(GROUP BY로 묶인 것 안에서 선택), ORDER BY(, AS(rename)
 * JOIN : 합친다
  > LEFT, RIGHT, FULL, CROSS(모두 한 번씩 JOIN)...
- UPDATE table_name SET column1 = value1, column2 = value2... WHERE condition;
- DELETE FROM table_name where condition;
 * WHERE 절을 빠뜨리면 전부다 삭제됨! 즉, DELETE는 조심해서 쓰던지, 아예 쓰질 말던지 해야 함
 * SQL injection에서 이걸로 개털릴 수 있음

## SQLAlchemy ORM
- class 또는 instance를 DB와 Mapping
- base : metadata처럼 mem. 에 있는 object를 관리하기 위함

```python
from sqlalchemy import create_engine
engine = create_engine("sqlite://", echo=True)
```

```python
from sqlalchemy.ext.declarative import declarative_base
```

```python
Base = declarative_base()
```

```python
from sqlalchemy import Column, Integer, String, ForeignKey
```

```python
class Artist(Base):
    __tablename__="Artist"
    id = Column(Integer, primary_key=True)
    name = Column(String)
    
    def __repr__(self):
        return "<T Artist(name='%s)>" % (self.name, )
    
```

```python
class Album(Base):
    __tablename__="Album"
    id = Column(Integer, primary_key=True)
    title = Column(String)
    artist_id = Column(Integer, ForeignKey('Artist.id'))
    
    def __repr__(self):
        return "<T Album(name='%s)>" % (self.title, )
    
```

```python
class Genre(Base):
    __tablename__="Genre"
    id = Column(Integer, primary_key=True)
    name = Column(String)
    
    def __repr__(self):
        return "<T Genre(name='%s)>" % (self.name, )
    
```

```python
class Track(Base):
    __tablename__="Track"
    id = Column(Integer, primary_key=True)
    title = Column(String)
    album_id = Column(Integer, ForeignKey('Album.id'))
    genre_id = Column(Integer, ForeignKey('Genre.id'))
    
    def __repr__(self):
        return "<T Track(name='%s)>" % (self.title, )
    
```

```python
Base.metadata.create_all(engine)
```
